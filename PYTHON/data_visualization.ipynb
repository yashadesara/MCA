```python
import pandas as pd
import matplotlib.pyplot as p
datapath = "data2.csv"
df = pd.read_csv(datapath)
```

```python
x = df['empid']
y = df['salary']
p.bar(x,y, label="employee data",color="lightgreen")
# p.barh(x,y, label="employee data",color="lightgreen")
p.xlabel("employee ids")
p.ylabel("employee salary")
p.title("Bar Example")
p.legend()
p.show()
```

```python
x2 = [1001,1003,1005,1007,1009,1011]
y2 = [1000,2000,3000,4000,5000,6000]

x3 = [1002,1004,1006,1008,1010,1012]
y3 = [2000,1000,5000,8000,1500,2500]

p.bar(x2,y2, label="sales dept",color="lightgreen")
p.bar(x3,y3, label="production dept",color="grey")

p.xlabel("employee ids")
p.ylabel("employee salary")

p.title("Bar Combined Example")
p.legend()
p.show()
```

```python
emp_ages = [22,43,32,56,34,38,50,10,30,56]
bins = [10,20,30,40,50,60]
p.hist(emp_ages,bins, label="employee data",color="lightgreen",alpha=0.9,rwidth=0.8)
p.xlabel("employee ids")
p.ylabel("employee salary")
p.title("Histogram Example")
p.legend()
p.show()
```

```python
slice = [50,20,15,15]
depts = ["sales","production","HR","finance"]
cols = ['lightgreen','lightblue','orange','grey']
p.pie(slice,labels=depts,colors=cols,startangle=90,explode=(0,0,0.2,0.1),
    shadow=True,autopct="%.2f%%")
p.title("Pie Exapmle")
p.legend()
p.show()
```

```python
years = ["2012","2013","2014","2015","2016","2017"]
profit = [5,7,8,7,7,5]

p.plot(years,profit,"lightgreen", label="profit")

p.xlabel("years")
p.xlabel("profit in Rs.")

p.title("Plot Example")
p.legend()
p.show()
```

```python
principal = 10000
interestRate = 0.05
years = 20
values = []

for i in range(years+1):
    values.append(principal)
    principal += principal * interestRate

p.plot(values, linewidth=30)
p.title("5% growth, compouned anually", fontsize='xx-large')
p.xlabel("years of compouned", fontsize='x-small')
p.ylabel('value of principal ($)')
p.show()
```

```python
p.rcParams['lines.linewidth'] = 5
p.rcParams['axes.titlesize'] = 25
p.rcParams['axes.labelsize'] = 25
p.rcParams['xtick.labelsize'] = 16
p.rcParams['ytick.labelsize'] = 16
p.rcParams['xtick.major.size'] = 7
p.rcParams['ytick.major.size'] = 7
p.rcParams['lines.markersize'] = 10
p.figure(1)
p.title("Example")
p.xlabel("x Label")
p.ylabel("y Label")
p.plot([1,2,3,5],[1,4,3,5])
p.show()
```
